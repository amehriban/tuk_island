```python
import numpy as np
import pygimli as pg
#import pybert as pb
from pygimli.physics import ert
#from pybert.importer.exportData import exportRes2dInv
from pygimli.physics.ert.importData import importRes2dInv
import matplotlib.pyplot as plt
import pygimli.meshtools as mt
from pygimli.frameworks import PriorModelling, JointModelling
from scipy.interpolate import interp1d
import pandas as pd
import pickle
from datetime import datetime
from glob import glob
```

```python
dir='//smb.isipd.dmawi.de/projects/p_subaqua/Mehriban/4pointlight/timelapse1/'
```

```python
DATA = [ert.load(file) for file in glob(dir+'*.tx0')]

tiemposs=['05/06/2024 13:06','05/06/2024 13:21','05/06/2024 13:36','05/06/2024 13:51'
                        ,'05/06/2024 14:06']

dates= [str(datetime.strptime(t, "%d/%m/%Y %H:%M")) for t in tiemposs]

tiempos= [datetime.fromisoformat(x) for x in dates]

for data in DATA:
    data['k'] = ert.createGeometricFactors(data, numerical=True)
    data['r'] = data['u'] / data['i']
    data['rhoa'] = data['k']*data['r']
    data['err']= ert.estimateError(data, relativeError=0.01) 
    print(data)

tl = ert.TimelapseERT(DATA, times=tiempos)
```

```python
tl.fullInversion()
tl.generateDataPDF()  # pseudosections
tl.generateModelPDF() # inversions
tl.generateRatioPDF(cMax=1.1)

tl.showAllModels()
```
